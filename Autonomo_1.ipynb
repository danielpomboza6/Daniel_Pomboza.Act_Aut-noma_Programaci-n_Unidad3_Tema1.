```python
class AnalizadorDatos:

    def __init__(self, datos):
        self.datos = datos

    def obtener_maximo(self):
        return max(self.datos)

    def obtener_minimo(self):
        return min(self.datos)

    def calcular_media(self):
        return sum(self.datos) / len(self.datos)

    def agregar_dato(self, nuevo_valor):
        self.datos.append(nuevo_valor)

    def mostrar_resumen(self):
        maximo = self.obtener_maximo()
        minimo = self.obtener_minimo()
        media = self.calcular_media()
        rango = maximo - minimo

        print("RESUMEN DE DATOS")
        print("Datos:", self.datos)
        print("Máximo:", maximo)
        print("Mínimo:", minimo)
        print("Media:", media)
        print("Rango:", rango)


# a) Lista de 10 valores
datos_iniciales = [7, 8, 10, 6, 9, 5, 8, 7, 10, 9]

# b) Crear instancia
analizador = AnalizadorDatos(datos_iniciales)

# c) Ejecutar métodos
analizador.mostrar_resumen()


# 3b) Ingreso dinámico de datos
print("\nIngresa más datos (escribe 'fin' para terminar):")

while True:
    entrada = input("Nuevo valor: ")

    if entrada.lower() == "fin":
        break

    try:
        numero = float(entrada)
        analizador.agregar_dato(numero)
    except:
        print("Ingresa un número válido")


# 3c) Mostrar nuevo resumen
print("\n NUEVO RESUMEN CON DATOS ACTUALIZADOS:")
analizador.mostrar_resumen()
```
